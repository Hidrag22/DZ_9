```python
#ЗАДАНИЕ 1

#Загружаем модуль для работы с регулярными выражениями
import re

#Функция проверки номера авто
#Принимает на вход номер авто
#Возвращает цифробуквенный номер и регион или сообщение о том, что номер не валиден
def check_car_id(car_id):
    #Русские буквы и похожие на них английский буквы, которые могут встречаться в номерах
    #Буквы сегда прописные
    rus = 'АВЕКМНОРСТУХ'
    eng = 'ABEKMHOPCTYX'
    #Выражение для поиска: 1 буква, 3 цифры, 2 буквы, 2-3 цифры
    #Буквы русские и английские допускаются в любом сочетании, контроль начала и конца строки
    regex_find = r'\A([' + rus + eng + r']{1})(\d{3})([' + rus + eng + r']{2})(\d{3}|\d{2})\Z'
    result = re.findall(regex_find, car_id)
    #Если номер валиден, то возвращаем отдельно номер и регион
    if result:
        return str('Номер ' + ''.join(result[0][:3]) + ' валиден. Регион: ' + result[0][3])
    return 'Номер не валиден'

#Проверка
car_id = 'A222BC96'
print(check_car_id(car_id))
```

```python
#ЗАДАНИЕ 2

#Загружаем модуль для работы с регулярными выражениями
import re

#Функция удаления повторов слов
#Принимает на вход строку, возвращает эту строку с удалёнными повторами слов
def check_repeat(some_string):
    #Определяем слово по границам \b...\b, состоящее из более одной любой буквы \w+ 
    #после слова (\b\w+\b) должен стоять пробел \s и затем оно же само (ссылка на себя через \1)
    #повторов должно быть быть более одного (\b\w+\b)(\s\1)+, в этом случае заменяем всё найденное на одно слово \1
    return re.sub(r'(\b\w+\b)(\s\1)+', r'\1', some_string)

#Проверка
some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений.'
print(check_repeat(some_string))
```

```python
#ЗАДАНИЕ 3

#Загружаем модуль для работы с регулярными выражениями
import re

#Функция принимает строку со словами и возвращает акроним
def acronim(some_words):
    #Определяем слово по границам \b...\b 
    #Первая позиция после начала слова - буква англ или рус алфавита ([a-zA-Zа-яА-ЯёЁ])
    #Затем могут идти буквы, либо сразу конец слова
    #В результате все слова заменяем на их первую букву, но остаются пробелы, цифры, знаки и т.д.
    firt_letters_from_words = re.sub(r'\b([a-zA-Zа-яА-ЯёЁ])[a-zA-Zа-яА-ЯёЁ]*\b', r'\1', some_words) 
    #Удаляем всё, что не является буквами, а затем переводим все буквы в прописные
    return (re.sub(r'[^a-zA-Zа-яА-ЯёЁ]', r'', firt_letters_from_words)).upper()

#Проверка
some_words = 'Информационные технологии'   
print(acronim(some_words))   
```

```python
#ЗАДАНИЕ 4

#Загружаем модуль для работы с регулярными выражениями
import re

#Функция принимает список email-адресов и выводит их распределение по доменным зонам
def emails_to_domain(emails):
    #Словарь с результатами
    result = {}
    #Перебираем емайлы по одному
    for email in emails:
        #Ищем любые символы после знака @
        domain = re.findall(r'(?<=@).+', email)
        #Если в словаре уже есть такой результат, то увеличиваем счётчик на 1
        if domain[0] in result:
            result[domain[0]] += 1
        #Если нет, тто создаём запись со счётчиком = 1
        else:
            result[domain[0]] = 1
    #Выводим результат на экран
    for one_str in result:
        print(one_str + ': ' + str(result[one_str]))
    return    


#Проверка
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']   
emails_to_domain(emails)
```

```python
#ЗАДАНИЕ 5

#Загружаем модуль для работы с регулярными выражениями
import re

#Функция для подсчёта количества слов, начинающихся на гласные и согласные буквы
def word_counter(some_text):
    #Счётчик для подсчёта количества слов, начинающихся на гласную букву
    result = {'vowels': 0, 'consonants': 0}
    #Гласные русского и английского языка
    vowels = 'аАоОуУэЭыЫяЯёЁюЮеЕиИaAeEiIoOuUyY'
    #Разбиваем текст на слова и перебираем их (слова могут начинаться только на буквы анг и рус алфавита)
    for word in re.findall(r'[a-zA-Zа-яА-ЯёЁ]+', some_text):
        #Если слово начинается с гласной буквы, то увеличиваем счётчик на 1
        if re.findall(r'^[' + vowels + ']', word):
            result['vowels'] += 1
        else:
            result['consonants'] += 1
    #Выводим результат
    print('Слов на гласные буквы: ' + str(result['vowels']))
    print('Слов на согласные буквы: ' + str(result['consonants']))
    return
    
#Проверка  
some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'  
word_counter(some_text)
```

```python
#ЗАДАНИЕ 6

#Загружаем модуль для работы с регулярными выражениями
import re

#Функция проверки номера сотового телефона
def phone_check(phone):
    #Проверяем есть ли в строке буквы, если есть, то сразу бракуем такой номер
    if re.findall(r'[a-zA-Zа-яА-ЯёЁ]', phone):
        return 'Номер не валиден'    
    #Удаляем всё, что не является цифрами
    digits = re.sub(r'[^\d]', '', phone)
    #Проверяем формат с международным кодом (11 цифр)
    #Должно первой цифрой стоять 7 или 8, а за ней ещё 10 цифр и конец строки \b
    if re.findall(r'^[78]\d{10}\b', digits):
        return re.sub(r'.(\d{3})(\d{3})(\d{2})(\d{2})', r'+7-\1-\2-\3-\4', digits)
    #Проверяем формат без международного кода (10 цифр)
    if re.findall(r'^\d{10}\b', digits):
        return re.sub(r'(\d{3})(\d{3})(\d{2})(\d{2})', r'+7-\1-\2-\3-\4', digits)
    return 'Номер не валиден'

phone = '123-456-789'    
print(phone_check(phone))   
```
